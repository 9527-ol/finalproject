```python
df = pd.read_sql_query('select * from post_info', con=connection_string)
df.head()
```

```python
df['words_per_comment'] = df['posts'].apply(lambda x: len(x.split())/50)
print(df.head())
```

```python
map1 = {"I": 0, "E": 1}
map2 = {"N": 0, "S": 1}
map3 = {"T": 0, "F": 1}
map4 = {"J": 0, "P": 1}

df['I-E'] = df['type'].astype(str).str[0]
df['I-E'] = df['I-E'].map(map1)

df['N-S'] = df['type'].astype(str).str[1]
df['N-S'] = df['N-S'].map(map2)

df['T-F'] = df['type'].astype(str).str[2]
df['T-F'] = df['T-F'].map(map3)

df['J-P'] = df['type'].astype(str).str[3]
df['J-P'] = df['J-P'].map(map4)

print(df.head(10))
```

```python
from nltk.tokenize import RegexpTokenizer
# tokenizer = RegexpTokenizer(r'[A-Za-z]+')
```

```python

tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
```

```python
df['Text'] = df.posts.map(lambda x:tokenizer.tokenize(x))
df
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
```

```python
df['Text'] = df.Text.map(lambda l: [stemmer.stem(word) for word in l])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(df.type)
encoded_y = label_encoder.transform(df.type)
```

```python
X =df.posts # df[['posts','words_per_comment']]
y = df.label
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# print(X_train.shape)
# print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# 2. instantiate the vectorizer
vect = CountVectorizer()  # lowercase=True worse
vect.fit(X_train)

# 4. transform training data
X_train_dtm = vect.transform(X_train)
X_train_dtm.toarray()
# summarize
print(vect.vocabulary_)
```

```python
print(X_train_dtm)
```

```python
# 4. transform testing data (using fitted vocabulary) into a document-term matrix
X_test_dtm = vect.transform(X_test)
X_test_dtm.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
# TfidfVectorizer that combines all the options of CountVectorizer and TfidfTransformer in a single model:
tfidf_transformer = TfidfTransformer()
# tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_dtm)
X_train_tfidf = tfidf_transformer.fit_transform(X_train_dtm)
X_train_tfidf.shape
```

```python
X_test_tf = tfidf_transformer.transform(X_test_dtm)
X_test_tf.shape
```

```python
# X_test_tfid = vectorizer.transform(X_test)
# X_test_tfid
```

```python
# 1. import
from sklearn.naive_bayes import MultinomialNB

# 2. instantiate a Multinomial Naive Bayes model
nb = MultinomialNB()
```

```python
nb.fit(X_train_tfidf, y_train)
```

```python
#make class predictions for X_test_dtm
y_pred_class = nb.predict(X_test_tf)
```

```python
# calculate accuracy of class predictions
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python

from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest

iris = load_iris()
iris
```

```python
X, y = iris.data, iris.target
X
```

```python
y
```

```python
pca = PCA(n_components=2)
pca
```

```python
selection = SelectKBest(k=1)
selection
```

```python
combined_features = FeatureUnion([("pca", pca), ("univ_select", selection)])
combined_features
```

```python
X_features = combined_features.fit(X, y).transform(X)
print("Combined space has", X_features.shape[1], "features")
```
